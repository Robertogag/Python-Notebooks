Crearemos una función llamada nuevo para agregar nombres de usuario y contraseñas a un diccionario, solo debes completar un renglón. Posteriormente añadiremos un usuario y responderemos un par de preguntas.
Una segunda actividad consistirá en utilizar un diccionario ya existente para buscar y visualizar información

```python
 dicci={"usuario1":"clave1","Ana María":"AnMa", "Juanito":"AMIG","Usuario12":"1212","Sofi":"contras"}
def nuevo(dicci=dicci):
  x=input("añada su usuario")
  y=input("añada su contraseña")
  dicci[x]=y #añade el valor "y" para una clave "x" en el dicci para añadir el usuario que agregue la nueva persona
  return dicci
```

```python
nuevo(dicci)
```

```python
print(dicci) #Imprime dicci
```

```python
nuevo() #Usando la función nuevo() agrega un nuevo usuario y contraseña
```
