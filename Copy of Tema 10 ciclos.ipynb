```python
#Veamos que pasa si el diccionario tiene listas en lugar de un solo valor
dic2={"Ana":[9,5],"Juan":[10,5],"Sofia":[8,1],"José":[9.5,0]}
for a,b in dic2.items(): #la primera variable siempre es la clave (key) y la segunda el valor
  print(a,"es la clave",b,"es el valor")
```

```python
#El dic3 nos muestra las calificaciones de matemáticas, español e inglés de los mismos 4 alumnos,
#Completa el ciclo for, para obtener en que materia le fue peor a cada alumno
dic3={"Ana":[9,10,8],"Juan":[10,9,10],"Sofia":[8,9,10],"José":[9.5,10,2]}
for a,b in dic3.items():
  if b.index(min(b))==0:
    c='matematicas'
  elif b.index(min(b))==1:
    c='español'
  else:
    c='Ingles'
  print('A',a,'le fue peor en:', c)
```

```python
Arreglo=np.array(dic3) #Convertimos la lista a un arreglo
```

```python
print(Arreglo)
```

```python
for a,b in enumerate(Arreglo):
  print(a,b)
```

```python
for x in np.nditer(Arreglo):
  print(x)#usando np.nditer con un arreglo
```
