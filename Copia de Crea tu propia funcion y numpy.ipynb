Ahora crearemos la función t, toma en cuenta lo siguiente:
1. Debes definir la función t usando def y la sintaxis vista en clase, debe tener un argumento (podemos llamarle x)
2. Dentro de la función debes de:

*   convertir x a string
*   luego convertirlo a lista
*   borrarle los elementos a partir del de indice 5
*   volver a juntar los elementos usando join
*   convertir x a float
*   terminar la función con return x

Esta función truncará el número a 4 cifras siempre que no sea mayor o igual a 10000


```python
def t(x):
  xstr=str(x)
  xlist =list(xstr)
  if x >=10000:
    return x
  xlist=xlist[:5]
  xstr=''.join(xlist)
  x=float(xstr)
  return x
```

```python
t(1.29292929) #utiliza tu función t para truncar el número 1.29292929 y ve que funcione
```

Ahora vamos a mejorar la función creando la función trunca:
Copia y pega la celda donde definiste la función t, cambiale el nombre a trunca y agrega otro argumento al lado de la x,
puedes ponerle y.

Ese segundo argumento será el número de cifras que deseas, dentro de tu función cambia el 5 por y+1.
Debes ponerle más 1 por que el punto decimal ocupa un lugar.
Después revisa que funcione la función escribiendo trunca(1.2929292929,7) debería darte 1.292929

```python
def trunca(x, y):
  xstr=str(x)
  xlist =list(xstr)
  if x>=y:
    return 'Error: x debe ser menor que y'
  if x>=10000:
    return x
  xlist=xlist[:y+1]
  xstr=''.join(xlist)
  x=float(xstr)
  return x
```

```python
trunca(1.2929292929,7)
```

Numpy

Es una librería (abreviatura de numeric python). Nos permite usar arreglos (matrices) y por lo tanto realizar operaciones que no se podrían hacer con listas

```python
import numpy as np #Usualmente se importa como np (para no tener que repetir la palabra numpy cada que usemos una de sus funciones)
```

Si tenemos dos listas y queremos dividir sus elementos no se puede directamente. Ejemplo:

```python
L1=[1,2,3]
L2=[4,4,9]
```

```python
L1/L2 #Veamos como da error al dividir dos listas
```

Con numpy podemos crear dos arreglos y luego dividirlos

```python
A1=np.array(L1) #La función array convierte una lista a un arreglo
print(A1)
```

```python
A2=np.array(L2)
print(A2) #Haz lo mismo con L2, llamale al arreglo A2
```

```python
A1/A2 #Divide A1/A2
```
