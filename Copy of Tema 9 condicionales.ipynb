Filtrado de arreglos y pandas

```python
nombres=np.array(["María","Ana","Sofía","José","Juan"])#Trabajaremos con las calificaciones de 5 alumnos, sus nombres y calificaciones están en los arreglos de numpy
calif=np.array([10,9.5,5,7.2,8])
```

```python
calif>8 #¿Qué alumnos tienen una calificación mayor a 8?
```

```python
nombres[calif>8]
```

Para usar and o or con arreglos se usa np.logical_and(cond,cond)

```python
nombres[np.logical_and(calif>=6,calif<9)]#¿Qué alumnos tienen una calificación aprobatoria (con 6 aprueban) pero menor a 9
```
