```python
lista= [[estado['población'], estado['estudiantes']] for estado in estados1.values()] #for estado in estados1.values(): Esto itera sobre los valores del diccionario estados1. En cada iteración, la variable estado toma el valor de uno de los elementos del diccionario (en este caso, un diccionario anidado).
```

```python
lista=[[estado['población'], estado['estudiantes']] for estado in estados1.values()]
print(lista)
```

```python
arreglo=np.array(lista) #arreglo numpy array de la lista anterior
```

```python
arreglo[:,1] #muestra los elementos correspondientes a estudiantes de cada estado
```

Hagamos una gráfica con barras dobles

```python
n = len(lista)
x = np.arange(n)
width = 0.25
plt.bar(x - width, arreglo[:,0], width=width, label='Población')
plt.bar(x, arreglo[:,1], width=width, label='Educación')
plt.xticks(x, estados1.keys())
plt.legend(loc='best') #
plt.show()

```
